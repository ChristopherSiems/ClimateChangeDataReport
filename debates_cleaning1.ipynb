```python
import pandas
import pycountry
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
countries = [pycountry.countries.lookup(country).alpha_3 if country != 'Turkey' else 'TUR' for country in pandas.read_csv('processed data/merged.csv')['country_or_area'].unique()]
data = pandas.read_csv('data/un-general-debates.csv')
data = data.loc[(data['year'] >= 1990) & (data['year'] <= 2013)].sort_values(by = 'year')
data.reset_index(drop = True, inplace = True)
data
```

```python
stemmer = PorterStemmer().stem
for i in range(len(data)):
  temp = pandas.DataFrame(data.iloc[i]).transpose()
  temp.reset_index(inplace = True, drop = True)
  tokens = [stemmer(token) for token in word_tokenize(temp.at[0, 'text']) if len(token) > 2 and token not in set(stopwords.words('english'))]
  temp.loc[0, 'tokens'] = ' '.join(tokens)
  print(temp)
  temp.to_csv('processed data/debates_cleaned1.csv', mode = 'a', header = False, index = False)

data = pandas.read_csv('processed data/debates_cleaned1.csv')
```
