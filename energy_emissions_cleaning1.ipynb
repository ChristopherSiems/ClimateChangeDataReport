```python
import pandas as pd

df = pd.read_csv("data/all_energy_statistics.csv")
gr_df = pd.read_csv("data/greenhouse_gas_inventory_data_data.csv")

display(df)
display(gr_df)
```

```python
# filters to wanted years
df = df.loc[(df['year'] > 1990) & (df['year'] < 2014)]
all_country = df['country_or_area'].unique()

# filters, renames countries with multiple names
gr_df = gr_df.loc[(gr_df['year'] > 1990) & (gr_df['year'] < 2014)]
gr_df['country_or_area'] = gr_df['country_or_area'].replace('United States of America', 'United States')
gr_df['country_or_area'] = gr_df['country_or_area'].replace('Czech Republic', 'Czechia')

gr_country = gr_df['country_or_area'].unique()

# only keeps countries that are overlapped between the two datasets
overlap = []
non = []

for country in gr_country:
    if country in all_country:
        overlap.append(country)
    else:
        non.append(country)
        
print(overlap)
print(non)

df = df.loc[df['country_or_area'].isin(overlap)]
df
```

```python
commodities = df['commodity_transaction'].unique()
#display(len(commodity_csv))
# over 1000 diff types of commodities recordes, this code counts how many of each there are
# serves as a start to pick good categories that describe clean energy
coms = []
nums = []
for com in commodities:
    coms.append(com)
    num = len(df.loc[df['commodity_transaction'] == com])
    nums.append(num)
    
coms_df = pd.DataFrame({
    "commodity": coms,
    "num of obs": nums
})

sorted_coms = coms_df.sort_values(by = 'num of obs', ascending = False)
display(sorted_coms)
sorted_coms.to_csv("processed data\coms.csv")
```

```python
gr_df['category'].unique()
# takes only the greenhouse gas data, could also do co2
gr_df = gr_df.loc[gr_df['category'] == 'greenhouse_gas_ghgs_emissions_including_indirect_co2_without_lulucf_in_kilotonne_co2_equivalent']
gr_df = gr_df[['country_or_area', 'year', 'value']]
```

```python
# lists of important commodities to research

clean_energy_com = [
    'Biogases - Final energy consumption',
    'Biogases - Production'
    'Direct use of solar thermal heat - Final energy consumption',
    'Electricity - total solar production',
    'Electricity - total wind production',
    'Direct use of geothermal heat - Final energy consumption',
    'Electricity - total hydro production'
]
fossil_energy_com = [
    'Liquified Petroleum Gas (LPG) - Final energy consumption',
    'Gas Oil/ Diesel Oil - Final energy consumption',
    'Motor Gasoline - Final energy consumption',
    'Hard coal - final energy consumption',
    'Natural Gas (including LNG) - final energy consumption','Heat - Final energy consumption',
    'Petroleum Coke - Final energy consumption','Electricity - total nuclear production',
]
misc_energy_com = [
    'Electricity - Gross demand',
    'Electricity - Gross production',
]

combined_com = clean_energy_com + fossil_energy_com + misc_energy_com
combined_com
```

```python
# filters data so only wanted data is in df, rearanges data nto more readable form
filter_df = df.loc[df['commodity_transaction'].isin(combined_com)]
filter_df = filter_df[['country_or_area', 'commodity_transaction', 'year', 'quantity']]
filter_df = filter_df.pivot_table(index=['country_or_area', 'year'], columns='commodity_transaction', values='quantity').reset_index()
filter_df.columns.name = None
filter_df
```

```python
# merges with gr gas df by country and year, keeps only data that have both
merged_df = pd.merge(filter_df, gr_df, on=['country_or_area', 'year'], how='inner')
merged_df.to_csv('processed data\merged.csv', index=False)
```
